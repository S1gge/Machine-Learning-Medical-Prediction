```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('cardio_train.csv', sep=';', index_col=0)
df.head()
```

```python
df.info()
```

```python
# Hur många är positiva för hjärt-kärlsjukdom och hur många ̈ar negativa? 
vc = df['cardio'].value_counts()
plt.pie(vc, autopct='%1.2f%%',  explode=[0.1,0], shadow=True)
plt.title('Proportion of people who tests positive or negative\n for cardiovascualr disease.')
plt.legend(loc='best', labels=[f'Negative: {df[df['cardio']==0]['cardio'].count()}',f'Positive: {df[df['cardio']==1]['cardio'].count()}'])
plt.show()
```

```python
# Hur stor andel har normala, ̈over normala och långt ̈over normala kolesterolvärden?

norm = df[df['cholesterol']==1]['cholesterol'].count()
above_norm = df[df['cholesterol']==2]['cholesterol'].count()
well_above_norm = df[df['cholesterol']==3]['cholesterol'].count()

vc = df['cholesterol'].value_counts()

plt.pie(vc,  autopct='%1.2f%%', explode=[0.04,0.1,0.1], shadow=True)
plt.title('Proportion of women and men who have normal, above normal \n and well above normal cholesterol levels.')
plt.legend(loc='best', labels=['Normal','Above normal', 'Well above normal'])
plt.show()

print(f'Number of people with normal cholesterol: {norm}')
print(f'Number of people with above normal cholesterol: {above_norm}')
print(f'Number of people with well above normal cholesterol: {well_above_norm}')
```

```python
# Hur ser ålders fördelningen ut?

fig, ax = plt.figure(figsize=(6,4), dpi=100), plt.axes()
sns.histplot(data=[a/365 for a in df['age']])
ax.set(xlabel='Age, years')
plt.show()

```

```python
# Hur stor andel röker?
vc = df['smoke'].value_counts()

plt.pie(vc, autopct='%1.2f%%', explode=[0.2,0], shadow=True)
plt.title('Proportion of women and men who smoke or not.')
plt.legend(loc='best', labels=["Don't smoke","Smoke"])
plt.show()
```

```python
# Hur ser vikt fördelningen ut?

fig, ax = plt.figure(figsize=(6,4), dpi=100), plt.axes()
sns.histplot(data=df['weight'], bins=80)
plt.title('Proportion of women and men who smoke or not.')
ax.set(xlabel='Weight, kg')
plt.show()
```

```python
# Hur ser längd fördelningen ut?

fig, ax = plt.figure(figsize=(6,4), dpi=100), plt.axes()
sns.histplot(data=df['height'], bins=80)
ax.set(xlabel='Age (years)')
plt.show()
```

```python
# Hur stor andel av kvinnor respektive män har hjärt-kärl sjukdom


df_pos = df[df['cardio']==1]
df_pos_g = df_pos['gender'].value_counts()


plt.pie(df_pos_g, autopct='%1.2f%%', explode=[0.1,0], shadow=True)
plt.title('Proportion of women and men who have cardiovascular disease.')
plt.legend(loc='best', labels=['Male','Female'])
plt.show()
```
